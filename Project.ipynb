```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_train['department'] = le.fit_transform(df_train['department'])
df_train['region'] = le.fit_transform(df_train['region'])
df_train['gender'] = le.fit_transform(df_train['gender'])
df_train['education'] = le.fit_transform(df_train['education'])
df_train['recruitment_channel'] = le.fit_transform(df_train['recruitment_channel'])
df_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_train[['no_of_trainings','age','previos_year_rating','length_of_service','avg_training_score']] = scaler.fit_transform(df_train[['no_of_trainings','age','previous_year_rating','length_of_service','avg_training_score']])
df_train
```

```python
#Splitting the data
from sklearn.model_selection import train_test_split
x = df_train.drop('is_promoted', axis=1)
y = df_train['is_promoted']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print("Shape of x_train: ", x_train.shape)
print("Shape of x_test: ", x_test.shape)
print("Shape of y_train: ", y_train.shape)
print("Shape of y_test: ", y_test.shape)
```

```python
from sklearn.impute import SimpleImputer

# Replace NaNs with the mean of each column (you can also use 'median' or 'most_frequent')
imputer = SimpleImputer(strategy='mean')
x_train = imputer.fit_transform(x_train)
```

```python
# Model Selection Training - Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=10000)
lr.fit(x_train, y_train)
```

```python
#Evaluation
y_pred_lr = lr.predict(x_test)
y_pred_lr
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred_lr))
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
```

```python
# MODEL  k-Nearest Neighbors (kNN)
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
y_pred_knn = knn.predict(x_test)
y_pred_knn
```

```python
print(accuracy_score(y_test, y_pred_knn))
print(confusion_matrix(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))
```

```python
# MODEL Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
y_pred_rf = rf.predict(x_test)
y_pred_rf
```

```python
print(accuracy_score(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
# MODEL  Naive Bayes
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)
```

```python
y_pred_nb = nb.predict(x_test)
y_pred_nb
```

```python
print(accuracy_score(y_test, y_pred_nb))
print(confusion_matrix(y_test, y_pred_nb))
print(classification_report(y_test, y_pred_nb))
```

```python
# MODEL MLP Classifier
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier()
mlp.fit(x_train, y_train)
```

```python
y_pred_mlp = mlp.predict(x_test)
y_pred_mlp
```

```python
print(accuracy_score(y_test, y_pred_mlp))
print(confusion_matrix(y_test, y_pred_mlp))
print(classification_report(y_test, y_pred_mlp))
```

```python
# MODEL GradientBoostingClassifier
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
gb.fit(x_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model = GradientBoostingClassifier(random_state=42)

param_grid = {
    'n_estimators': [100, 150],         # number of boosting stages
    'learning_rate': [0.05, 0.1],       # step size shrinkage
    'max_depth': [3, 4, 5],             # depth of individual trees
    'subsample': [0.8, 1.0]             # fraction of samples used for fitting
                }
```

```python
grid = GridSearchCV(
      estimator=model,
      param_grid=param_grid,
      scoring='f1',
      cv=3,
      verbose=2,
      n_jobs=--1
)

grid.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def get_metrics(model_name, y_true, y_pred):
    return {
            "Model": model_name,
            "Accuracy": accuracy_score(y_true, y_pred),
            "Precision": precision_score(y_true, y_pred),
            "Recall": recall_score(y_true, y_pred),
            "F1-Score": f1_score(y_true, y_pred)
}
```

```python
results = []

# Logistic Regression
results.append(get_metrics("Logistic Regression", y_test, y_pred_lr))

# k-Nearest Neighbors
results.append(get_metrics("k-NN", y_test, y_pred_knn))

# Random Forest
results.append(get_metrics("Random Forest", y_test, y_pred_rf))

# Naive Bayes
results.append(get_metrics("Naive Bayes", y_test, y_pred_nb))

# MLP Classifier
results.append(get_metrics("MLP Classifier", y_test, y_pred_mlp))

# Gradient Boosting
results.append(get_metrics("Gradient Boosting", y_test, y_pred_gb))
```

```python
comparison_df = pd.DataFrame(results)
comparison_df.set_index("Model", inplace=True)
comparison_df = comparison_df.sort_values(by="F1-Score", ascending=False)
comparison_df.style.background_gradient(cmap="YlGnBu", axis=1)
```
